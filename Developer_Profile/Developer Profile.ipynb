```python
# hide
responses_num = df['Country'].value_counts().head(10)
# responses_num.values
```

```python
plt.rcParams['figure.dpi'] = 100
sns.set(style="whitegrid")

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(responses_num.index, responses_num.values)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 12, fontweight ='bold',
             color ='grey')

# Add plot title
ax.set_title('Top 10 Countries', fontsize = 14,
             loc ='left', )
 
# Add Text watermark
fig.text(0.9, 0.15, f'{responses_num.values.sum()} responses', fontsize = 12,
         color ='black', ha ='right', va ='bottom',
         alpha = 0.7)
 
# Show Plot
plt.show()
```
